```python
import matplotlib.pyplot as plt
import pandas as pd
nobel_prizes = pd.read_csv('C:/Users/Admin/Curso Data Science/nobelPrizes.csv', index_col="Year")
nobel_prizes.head()
```

```python
categories = nobel_prizes['Category'].unique()
categories
```

```python
#Se extrae una serie para cada categoria
chemistry = nobel_prizes[nobel_prizes['Category']=='Chemistry']
literature = nobel_prizes[nobel_prizes['Category']=='Literature']
medicine = nobel_prizes[nobel_prizes['Category']=='Medicine']
peace = nobel_prizes[nobel_prizes['Category']=='Peace']
physics = nobel_prizes[nobel_prizes['Category']=='Physics']
economics = nobel_prizes[nobel_prizes['Category']=='Economics']
```

```python
#Testeando
chemistry_countries = chemistry['Birth Country'].value_counts().reset_index()
chemistry_countries
```

```python
#Se cuentan los países por categoría 
literature_countries = literature['Birth Country'].value_counts().reset_index()
medicine_countries = medicine['Birth Country'].value_counts().reset_index()
peace_countries= peace['Birth Country'].value_counts().reset_index()
physics_countries = physics['Birth Country'].value_counts().reset_index()
economics_countries = economics['Birth Country'].value_counts().reset_index()
```

```python
fig, ax = plt.subplots(2,3,figsize=(15, 10))
plt.subplots_adjust(hspace = 0.7)

#Gráficos
ax[0,0].bar(chemistry_countries['index'].iloc[0:5], chemistry_countries.iloc[0:5,1], label='Chemistry')
ax[0,1].bar(literature_countries['index'].iloc[0:5], literature_countries.iloc[0:5,1],label='Literature')
ax[0,2].bar(medicine_countries['index'].iloc[0:5], literature_countries.iloc[0:5,1],label='Medicine')
ax[1,0].bar(peace_countries['index'].iloc[0:5], peace_countries.iloc[0:5,1], label='Peace')
ax[1,1].bar(physics_countries['index'].iloc[0:5], physics_countries.iloc[0:5,1], label='Physics')
ax[1,2].bar(economics_countries['index'].iloc[0:5], economics_countries.iloc[0:5,1], label='Economics')

#Básicamente para rotar los ticks
ax[0,0].set_xticklabels(chemistry_countries['index'].iloc[0:5],rotation=90)
ax[0,1].set_xticklabels(literature_countries['index'].iloc[0:5],rotation=90)
ax[0,2].set_xticklabels(medicine_countries['index'].iloc[0:5],rotation=90)
ax[1,0].set_xticklabels(peace_countries['index'].iloc[0:5], rotation=90)
ax[1,1].set_xticklabels(physics_countries['index'].iloc[0:5], rotation=90)
ax[1,2].set_xticklabels(economics_countries['index'].iloc[0:5], rotation=90)

#Legends de cada gráfico
ax[0,0].legend()
ax[0,1].legend()
ax[0,2].legend()
ax[1,0].legend()
ax[1,1].legend()
ax[1,2].legend()
#Mostrar todos los gráficos
plt.show()
```
